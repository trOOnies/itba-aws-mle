# Data split

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
os.chdir("..")
os.getcwd()
```

## Data

Detailed info on **flight arrivals and delays for U.S. airports**, categorized by carriers.

It includes metrics such as the number of arriving flights, delays over 15 minutes, cancellation and diversion counts, and the breakdown of delays attributed to carriers, weather, NAS (National Airspace System), security, and late aircraft arrivals.

```python
df = pd.read_csv("data/complete.csv")
df
```

## Split

Hacemos un split 80-10-10 como es usual en el mundo de Machine Learning

```python
df_train, df_val_test = train_test_split(df, test_size=0.20)
df_val, df_test = train_test_split(df_val_test, test_size=0.50)
```

```python
print(df_train.shape)
print(df_val.shape)
print(df_test.shape)
```

```python
df_train.to_csv("data/train.csv", index=False)
df_val.to_csv("data/val.csv", index=False)
df_test.to_csv("data/test.csv", index=False)
```

```python
del df_val_test, df_test
```

```python
df_train_val = pd.concat((df_train, df_val))
```

```python
del df_train, df_val
```
